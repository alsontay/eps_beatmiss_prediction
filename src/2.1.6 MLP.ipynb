```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")

X_train = pd.read_csv("../data/train.csv")
X_test = pd.read_csv("../data/test.csv")
y_train = pd.read_csv("../data/train_labels.csv")
y_test = pd.read_csv("../data/test_labels.csv")
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
from tensorflow.keras.constraints import MaxNorm
import tensorflow as tf

tf.random.set_seed(100)
# create sequential multi-layer perceptron
nn_model = Sequential()
#1st input layer layer (1st hidden layer)
nn_model.add(Dropout(0.3))
nn_model.add(Dense(30, activation='relu', kernel_constraint=MaxNorm(3)))
#3rd layer (2nd hidden layer)
nn_model.add(Dense(30, activation='relu', kernel_constraint=MaxNorm(3)))
nn_model.add(Dense(30, activation='relu', kernel_constraint=MaxNorm(3)))
nn_model.add(Dense(30, activation='relu', kernel_constraint=MaxNorm(3)))
nn_model.add(Dense(30, activation='relu', kernel_constraint=MaxNorm(3)))
nn_model.add(Dense(30, activation='relu', kernel_constraint=MaxNorm(3)))
nn_model.add(Dense(30, activation='relu', kernel_constraint=MaxNorm(3)))
nn_model.add(Dense(30, activation='relu', kernel_constraint=MaxNorm(3)))
# nn_model.add(Dense(10, activation='relu', kernel_constraint=MaxNorm(2)))
# nn_model.add(Dense(10, activation='relu', kernel_constraint=MaxNorm(2)))
# nn_model.add(Dense(10, activation='relu', kernel_constraint=MaxNorm(2)))
# nn_model.add(Dense(10, activation='relu', kernel_constraint=MaxNorm(2)))
#output layer
nn_model.add(Dense(1, activation='sigmoid'))

nn_model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['acc'])

nn_model.fit(X_train, y_train.values.ravel(), epochs=200)
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score, f1_score

y_pred = nn_model.predict(X_train) 
y_pred[y_pred <= 0.5] = 0
y_pred[y_pred > 0.5] = 1
acc = accuracy_score(y_train, y_pred)
print(f"train accuracy: {acc}")

y_pred = nn_model.predict(X_test) 
y_pred[y_pred <= 0.5] = 0
y_pred[y_pred > 0.5] = 1
acc = accuracy_score(y_test, y_pred)
print()
print(f"test accuracy: {acc}")

print()
y_pred_proba = nn_model.predict(X_train)

roc_auc = roc_auc_score(y_train, y_pred_proba)
print(f"train roc auc score: {roc_auc}")

y_pred_proba = nn_model.predict(X_test)

roc_auc = roc_auc_score(y_test, y_pred_proba)
print()
print(f"test roc auc score: {roc_auc}")

print()
y_pred = nn_model.predict(X_train) 
y_pred[y_pred <= 0.5] = 0
y_pred[y_pred > 0.5] = 1
f1 = f1_score(y_train, y_pred)
print(f"train f1 score: {f1}")

y_pred = nn_model.predict(X_test) 
y_pred[y_pred <= 0.5] = 0
y_pred[y_pred > 0.5] = 1
f1 = f1_score(y_test, y_pred)
print()
print(f"test f1 score: {f1}")
```
